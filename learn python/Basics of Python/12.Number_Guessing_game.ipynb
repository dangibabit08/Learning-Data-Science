```python
import random
```

```python
random.randint(1,100)
```

```python
right_number=random.randint(1,100)
guess_number=int(input("Guess the correct number:"))
count=1
while guess_number != right_number:
    if guess_number<right_number:
        print("Guess higher")
    else:
        print("guess lower")
    guess_number=int(input("Again Guess the Number:"))
    count+=1
print("you guess the correct number is:",guess_number)    
print("total attempt is:",count)
```

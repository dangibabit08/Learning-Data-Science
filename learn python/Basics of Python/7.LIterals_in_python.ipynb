```python
#string literals

string='this is a python.'
strings="This is also a python."
char="C"
multiline_string="this is a multiline string with more than one line code."
unicode=u"\U0001f600\U0001F606\U0001F923"
raw_string=r"raw\n strings."
print(string)
print(strings)
print(char)
print(multiline_string)
print(unicode)
print(raw_string)
```

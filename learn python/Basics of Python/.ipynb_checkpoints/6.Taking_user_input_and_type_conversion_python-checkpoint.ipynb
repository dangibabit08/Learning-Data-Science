```python
#type conversion
#Implicit Conversion
#Explicit Conversion
```

```python
#Implicit conversion------> python automatically convert the data types
a=4  #int
b=4.56 #float
c=a+b
print(c)
print(type(c))
```

```python
#Explicit conversion ----->you manually convert to another data types.

first_number=int(input("enter the first number:"))
second_number=int(input("enter the second number:"))
sum=first_number+second_number
print(f"The sum of {first_number} and {second_number} is :{sum}")
```

```python
#Basic Types
#integer

print(4)
print(1e308)
print(1e309)    # this is the range of the int 
```

```python
#float

print(5.24)
print(1.7e308)
print(1.7e309)  #this is the range of float
```

```python
#boolean

print(True)
print(False)
```

```python
#complex
print(3+5j)
```

```python
#string

print('babit')
print("babit")
print("""babit""")
```

```python
#Container Types
#list  ----> they are like an arrray

print([1,2,3,4])
```

```python
#tuples

print((1,2,3,4))
```

```python
#sets

print({1,2,3,4})
```

```python
#dictionary ----> key-value pair

print({"Name":"Babit Dangi","Age": 24, "Gender":"Male"})
```

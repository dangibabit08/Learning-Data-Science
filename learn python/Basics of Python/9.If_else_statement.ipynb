```python
email=input("enter the email address")

if '@'  in email and '.' in email:
    password=input("enter the password")
    
    if email=="dangibabit@gmail.com" and password=="1234":
       print("you have successfully login the app")
    elif email=="dangibabit@gmail.com" and password != "1234":
       print("Incorrect password try again")
       password=input("enter the password again")
       if password=="1234":
           print("We have enter the right password welcome")
       else:
           print("incorrect password")
    else: 
       print("enter the correct email")

else:
  print("enter the right email")

```
